```python
import pandas as pd

df = pd.read_csv("../data/processed/epl_loaded.csv")

print(df.shape)

columns_to_keep = [
    "Date",
    "HomeTeam",
    "AwayTeam",
    "FTHG",
    "FTAG",
    "FTR",
    "HS",
    "AS",
    "HST",
    "AST",
    "HC",
    "AC",
    "HF",
    "AF"
]

df = df[columns_to_keep]

print(df.isnull().sum())

df = df.dropna()

df["Date"] = pd.to_datetime(
    df["Date"],
    format="%d/%m/%Y"
)

print(df.shape)

df.to_csv("../data/processed/epl_cleaned.csv", index=False)
```
