```python
import pandas as pd

df = pd.read_csv("../data/processed/epl_cleaned.csv")
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

df["TotalGoals"]= df["FTHG"] + df["FTAG"]
df["GoalDifference"] = df["FTHG"] - df["FTAG"]

#Rolling averages of last 10 games
df["HomeGoalsAvg"] = df.groupby("HomeTeam")["FTHG"].transform(lambda x: x.shift(1).rolling(10).mean())
df["AwayGoalsAvg"] = df.groupby("AwayTeam")["FTAG"].transform(lambda x: x.shift(1).rolling(10).mean())
df["HomeShotsOnTargetAvg"] = df.groupby("HomeTeam")["HST"].transform(lambda x: x.shift(1).rolling(10).mean())
df["AwayShotsOnTargetAvg"] = df.groupby("AwayTeam")["AST"].transform(lambda x: x.shift(1).rolling(10).mean())
df = df.dropna()

print(df.head())
df.to_csv("../data/processed/epl_features.csv", index=False)
```
